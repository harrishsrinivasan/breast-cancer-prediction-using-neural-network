```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

# DATA COLLECTION AND PROCESSING

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
```

```python
#loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_data.data, columns = breast_cancer_data.feature_names)
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_data.target
```

Splitting the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

# **BUILDING THE NEURAL NETWORK**

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras import layers
```

```python
# setting up the layers of neural network

model = keras.Sequential([
    keras.Input(shape=(30,)),
    layers.Flatten(),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compliting the neural network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network

history = model.fit(X_train, Y_train, validation_split=0.1, epochs=20)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

accuracy of the model in test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
